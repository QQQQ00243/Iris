# Prepare the data for ML learning Algorithms

## Do stratified sampling based on the iris category

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(irising, irising["cat"]):
    strat_train_set = irising.loc[train_index]
    strat_test_set = irising.loc[test_index]
```

```python
iris_train = strat_train_set.iloc[:, :-2]
iris_cat = strat_train_set["cat"]

iris_test = strat_test_set.iloc[:, :-2]
iris_cat_test = strat_test_set["cat"]
```

# Decision Tree Classification

```python
from sklearn import tree
from sklearn.model_selection import cross_val_score

clf = tree.DecisionTreeClassifier()
cross_val_score(clf, iris_train, iris_cat, cv=3, scoring="accuracy")
```

# Error Analysis

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

iris_train_pred = cross_val_predict(clf, iris_train, iris_cat, cv=3)
conf_mx = confusion_matrix(iris_cat, iris_train_pred)
conf_mx
```

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()
```

The confusion matrix looks fairly good. But 2 elements in class 1 are mistakenly classified into class 2 and 4 elements in class 2 into class 1.

Let's focus on these errors.

```python
cl1, cl2 = 1, 2

# iris_ij, i represents the actual class and j the predicted
iris_11 = iris_train[(iris_cat == cl1) & (iris_train_pred == cl1)]
iris_12 = iris_train[(iris_cat == cl1) & (iris_train_pred == cl2)]
iris_21 = iris_train[(iris_cat == cl2) & (iris_train_pred == cl1)]
iris_22 = iris_train[(iris_cat == cl2) & (iris_train_pred == cl2)]

index_false = iris_12.index.to_list() + iris_21.index.to_list()
iris_train_true = strat_train_set.drop(index_false)
```

```python
feature_names = iris.feature_names
n = len(feature_names)
combinations = []


for i in range(n):
  for j in range(i+1, n):
    combinations.append([feature_names[i], feature_names[j]])
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

for i in range(6):
  temp = combinations[i]
  x, y = temp[0], temp[1]

  sns.scatterplot(data=iris_train_true, ax=axes[i//2, i%2],
                  x=x, y=y, hue='cat_name')
  sns.scatterplot(data=iris_12, x=x, y=y, ax=axes[i//2, i%2],
                  marker="x", color="r", s=100, label="iris_12")
  sns.scatterplot(data=iris_21, x=x, y=y, ax=axes[i//2, i%2],
                  marker="+", color="r", s=100, label="iris_21")
```

We notice that the mistakenly classified instances are on the verge of vericolor and virginica classes while the recall and precise rate of setosa are 100%, which means it is harder to distinguish vericolor and virginica flowers.

# RandomizedSearchCV

```python
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV


random_grid = {"max_depth": [3, None],
              "max_features": randint(1, 9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}
```

```python
clf = tree.DecisionTreeClassifier()
rnd_search = RandomizedSearchCV(
    estimator=clf, param_distributions = random_grid, 
    n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = 4
    )
rnd_search.fit(iris_train, iris_cat)
```

```python
best_random = rnd_search.best_estimator_
cross_val_score(best_random, iris_train, iris_cat, cv=3, scoring="accuracy")
```

# Final Result

```python
from sklearn.metrics import accuracy_score

iris_cat_pred = best_random.predict(iris_test)
accuracy_score(iris_cat_pred, iris_cat_test)
```
